# Array Inspection

```python
import numpy as np
```

```python
a = np.array(20)
print(a)
print(type(a))
print(np.size(a)) # it wil give you size
print(np.ndim(a))
print(np.shape(a))
```

```python
a = np.array([20,10,30])
print(a)
print(type(a))
print(np.size(a)) # it wil give you size
print(np.ndim(a))
print(np.shape(a))
```

```python
a = np.array([[20,10,30],
             [20,10,30]])
print(a)
print(type(a))
print(np.size(a)) # it wil give you size
print(np.ndim(a))
print(np.shape(a))
```

```python
a = np.array([[[20,10,30],
             [20,10,30],
             [20,10,30]]])
print(a)
print(type(a))
print(np.size(a)) # it wil give you size
print(np.ndim(a))
print(np.shape(a))
```

```python
a = np.array([[[20,10,30],
             [20,10,30],
             [20,10,30]],
             
             [[20,10,30],
             [20,10,30],
             [20,10,30]]])
print(a)
print(type(a))
print(np.size(a)) # it wil give you size
print(np.ndim(a))
print(np.shape(a))
```

# Array Mathematics

```python
a = [10,20,30,40,50]
print(np.min(a))
print(np.max(a))
print(np.mean(a))
print(np.median(a))
print(np.std(a))
```

```python
a = 10 
b = 20 
print(np.sum([a,b]))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
np.sum([10,20,30,40,50])
```

# Reshaping 

```python
a = np.arange(24, dtype = float)
print(a)
print(np.ndim(a))
print()
b = a.reshape(6,4)
print(b)
print(np.ndim(b))
print()
c = a.reshape(2,3,4)
print(c)
print(np.ndim(c))
```

# Indexing And Slicing

```python
a = [10,20,30,40]
# 10 
# a[index]
a[0]
```

```python
# 30
a[2]
```

```python
a = np.array([10,20,30,40])
# 10
a[0]
```

```python
a[2]
```

```python
a = np.array([10,20,30,40])
# [10 20]
# a[start:end+1:step]
print(a[0:2:1])
print(a[0:2])
print(a[:2])
```

```python
a = np.array([10,20,30,40])
# [10,40]
a[0::3]
a[0:4:3]
a[::3]
```

```python
a = np.array([[10,20,30,40],
              [11,21,31,41]])
# indexing
# a[rows, columns]
a[0,3]

```

```python
a = np.array([[10,20,30,40],
              [11,21,31,41]])
# slicing
# a[rows, columns]
# a[rstart:rend+1:step, cstart:send+1:cstep]
a[0:0+1:1,1:3:1]

# [[20,30],
# [21,31]]
a[0:2:1,1:3:1]
```

```python
a = np.array([[10,20,30,40],
              [11,21,31,41],
             [12,22,32,42]])

# [[10,40],
# [12,42]]
a[0::2,0::3]
a[0:3:2,0:4:3]
```
