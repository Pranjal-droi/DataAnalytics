# Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(0,10,20)
plt.figure(figsize=(6,4))
ax1=plt.subplot()
ax1.plot(a,linewidth=1,linestyle='--',
         color='r',marker='o',alpha=0.5)
plt.xlabel('no. of point')
plt.ylabel('data values')
plt.title('lineplot')
plt.grid(True)
plt.savefig('lineplot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.linspace(0,10,20)
plt.figure(figsize=(6,4))
ax1=plt.subplot()
ax1.plot(a,linewidth=1,linestyle='--',
         color='r',marker='o',alpha=0.5)
plt.xlabel('no. of point', fontweight='bold', fontsize=14)
plt.ylabel('data values', fontweight='bold', fontsize=14)
plt.title('lineplot', fontweight='bold', fontsize=14)
plt.grid(True)
plt.savefig('lineplot.png')
plt.show()
```
